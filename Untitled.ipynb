## Independent And Dependent Features

```python
bostan_df.iloc[:,:-1]
```

```python
# Independent or Dependent Features
X = bostan_df.iloc[:,:-1]
Y  = bostan_df.iloc[:,-1]
```

```python
##Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size =0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test= scaler.transform(X_test)
```

## Here is the qustion why we use fit_transform( ) on training data but transform( ) on the test data?
The fit method is calculating the mean and variance of each of the features present in our data.The transform method is transforming all the features using the respective mean and variance.

Now we want scaling to be applied to our test data too and at the same time do not want to be biased with our model. we want our test data to be a completley new and surprise set for our model.The transform method help us in this case.

__transform()__

using the transform() method we can use the same mean and variance as it is calculate from our traning data to transform our test data. Thus the parameters learned by our model using the traning data will help us to transfrom our test data.

## Model Traning

```python
from sklearn.linear_model import LinearRegression
```

```python
Regression = LinearRegression()
```

```python
Regression.fit(X_train,y_train)
```

```python
## print the Coefficients and the intercpet
print(Regression.coef_)
```

```python
print(Regression.intercept_)
```

```python
## on which Paranmters the model has been trained
Regression.get_params()
```

```python
## Predication wiht test Data 
reg_pred = Regression.predict(X_test)
```

```python
#plot a scatter plot for the predication
plt.scatter(y_test,reg_pred)
```

```python
# Residualsy Meant that coutning the error
Residualsy = y_test- reg_pred
```

```python
#ploting the residuals
sns.displot(Residualsy,kind ="kde")
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error 
```

```python
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
```

## R Square and adjusted R square

## R Square 

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
print(score)
```

## Adjusted R Square

```python
1 -(1-score) * (len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
